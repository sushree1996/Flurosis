```python
# Libraries 
from sklearn import datasets 
import pandas as pd
import seaborn as sns
import numpy as np
from statsmodels.stats import weightstats as stests
import statsmodels.api as sm
from statsmodels.formula.api import ols 
import warnings
warnings.simplefilter('ignore')
```

```python
df = pd.read_csv('chisquare.csv')
df.head()
```

```python
# summarise number of missing values again
100*(df.isnull().sum()/len(df.index))
```

```python
# Now we can check Interaction Effect between Class and Bin_Petal_Width through Chi Square Test
#Received information and Frequency of visit
# First Create Contigency Table
contingency_table = pd.crosstab(df['306.Have you received information about ill effects about excessive fluoride in water from government?'],df['508. What is the frequency for visiting the dentist?'])
print(contingency_table)
```

```python
# Chi Square Test
from scipy import stats
chi2_stat,p_val,dof,ex = stats.chi2_contingency(contingency_table)
print("CHI-SQUARE TEST VALUES")
print("Chi Square Value : ",chi2_stat)
print("Degree of Freedom : ",dof)
print("P Value : ", p_val)
print("Contigency Table")
ex
```
